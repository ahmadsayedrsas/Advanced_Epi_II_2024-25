# Exercise: Hierarchical Cluster Analysis
You are provided with a dataset containing demographic indicators for 36 European countries. These include variables such as:

- birth and death rates  
- male and female life expectancy  
- total population and population density

Your goal is to uncover clusters of countries that share similar demographic profiles. To do this, proceed as follows:

- Design and implement a **machine learning pipeline** that prepares your data for clustering.
- Use **PCA** to reduce the dimensionality of the data and support your interpretation.
- Apply **hierarchical clustering** to identify meaningful groups of countries.
- Visualize your results using a **dendrogram**, and analyze the outcome.
- Investigate and compare **different linkage methods** (`ward`, `complete`, `average`, etc.). Justify your final choice.
- Characterize each resulting cluster: What do the countries in each group have in common? How do they differ from others?


```python
import pandas as pd

df = pd.read_csv("europe_demo_cleaned.csv")
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
import matplotlib.pyplot as plt
import seaborn as sns


features = df.drop(columns=["country_name"])
scaler = StandardScaler()
scaled_data = scaler.fit_transform(features)

pca = PCA(n_components=2)
pca_components = pca.fit_transform(scaled_data)

pca_df = pd.DataFrame(pca_components, columns=["PC1", "PC2"])
pca_df["country"] = df["country_name"]

linkage_methods = ["ward", "complete", "average"]
linkage_matrices = {method: linkage(scaled_data, method=method) for method in linkage_methods}

fig, axes = plt.subplots(1, 3, figsize=(20, 6))
for ax, method in zip(axes, linkage_methods):
    dendrogram(linkage_matrices[method], labels=df["country_name"].values, ax=ax, leaf_rotation=90)
    ax.set_title(f"{method.capitalize()} Linkage")
    ax.set_ylabel("Distance")
plt.tight_layout()
plt.show()

```

We can choose ward linkage as it produces well-seperated homogenous clusters.

```python
ward_linkage = linkage_matrices["ward"]
plt.figure(figsize=(12, 6))
dendrogram(ward_linkage, labels=df["country_name"].values, leaf_rotation=90, color_threshold=10)
plt.title("Hierarchical Clustering Dendrogram (Ward Linkage)")
plt.ylabel("Distance")
plt.axhline(y=10, color='r', linestyle='--', label='Cut-off for 3 clusters')
plt.legend()
plt.tight_layout()
plt.show()

cluster_labels = fcluster(ward_linkage, t=3, criterion='maxclust')
df["cluster"] = cluster_labels
pca_df["cluster"] = cluster_labels

plt.figure(figsize=(10, 7))
sns.scatterplot(data=pca_df, x="PC1", y="PC2", hue="cluster", palette="tab10", s=100)
for i in range(len(pca_df)):
    plt.text(pca_df["PC1"][i]+0.05, pca_df["PC2"][i], pca_df["country"][i], fontsize=8)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(title="Cluster")
plt.tight_layout()
plt.show()

cluster_summary = df.groupby("cluster").mean(numeric_only=True)
print(cluster_summary.round(2))
```

Cluster 1 mostly includes Eastern European counntries and is charachterized by having a higher death rate and lower life expectancy, especially for men.
Cluster 2 includes countries with slightly higher birth rates and high life expectancy.
Cluster 3 includes only Malta and its birth and death rates are noticably lower than the other clusters.
